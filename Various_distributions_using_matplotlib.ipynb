##  Various discrete distributions in Statistics that are useful in Data Analysis and Data Science
##  Each having its own importance and variations


Discrete Uniform Distribution: It is the discrete probability distribution used when there is a finite number of outcomes

Ex: Rolling a die; the possible outcomes of the upper face of the die is (1,2,3,4,5,6)

Bernoullis Distribution: It is the discrete probability distribution that describes outcomes of a single trail with two possible outcomes success(1) or failure(0)

Ex: Probability of getting head in single toss of coin.

```python
##Generating data from bernoullis  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.binomial(n=1,p=0.5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Binomial Distribution: Binomial Distribution is a discrete probability distribution that describes the number of successes in n independent Bernoulli trials, where each trial has the same probability of success,p.

Ex: The probability of getting head, when a fair coin is tossed 8 times.


```python
#Generating data from binomial distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=np.random.binomial(n=10,p=0.6,size=2000)
plt.data=sns.histplot(data,kde=True)
plt.show()

```

Poisson Distribution: It is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space, given that these events happen independently and with a known constant rate.

Ex: Number of accidents in a day

```python
#Generating data from poisson distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.poisson(lam=5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Geometeric Distribution: Geometric Distribution is a discrete probability distribution that models the number of trials needed to achieve the first success in a series of independent Bernoulli trials, where each trial has a constant probability of success p.

Ex:Number of machines tested until finding the first defective one.

```python
##Generating data from geometeric  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.geometric(p=0.5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Hyper Geometeric Distribution:The Hypergeometric Distribution is a discrete probability distribution that describes the probability of k successes in
n draws from a finite population without replacement. It is used when the sample is drawn from a finite population and the sampling is done without replacement, meaning that the probability of success changes with each draw.

Ex: Lottery Drawing; If there are 100 tickets, 10 of which are winning tickets, and you draw 5 tickets, the Hypergeometric Distribution describes the probability of drawing exactly k winning tickets.

```python
##Generating data from hyper geometeric  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Parameters for the Hypergeometric distribution
N = 50        # Total population size
K = 20        # Number of successes in the population
n = 10        # Number of draws

# Generate data
data = np.random.hypergeometric(ngood=K, nbad=N-K, nsample=n, size=10000)

# Calculate skewness
data_skewness = skew(data)

# Plot the histogram with KDE
sns.histplot(data, kde=True, color="skyblue", bins=20)
plt.title(f"Hypergeometric Distribution (Skewness: {data_skewness:.2f})")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

# Print skewness value
print(f"Skewness of the data: {data_skewness:.2f}")
```

Negative Binomial Distribution:It is a discrete probability distribution that describes the number of successes in a sample drawn from a finite population, where each draw is without replacement. It is typically used in situations where the total population size, the number of successes in the population, and the sample size are known, and it is not appropriate to assume that the draws are independent.

Ex: If a factory produces 1,000 items, and 200 of them are defective, the Hypergeometric Distribution can model the probability of finding exactly
k defective items in a sample of 50.

```python
##Generating data from negative binomial  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Parameters
N = 50   # Total population size
K = 20   # Number of successes in the population
r = 5    # Target number of successes
size = 1000  # Number of samples

def negative_hypergeometric(N, K, r, size):
    data = []
    for _ in range(size):
        successes = 0
        failures = 0
        population = np.random.permutation([1] * K + [0] * (N - K))
        for item in population:
            if successes == r:
                break
            if item == 1:
                successes += 1
            else:
                failures += 1
        data.append(failures)
    return np.array(data)
# Generate data
data = negative_hypergeometric(N, K, r, size)

# Calculate skewness
data_skewness = skew(data)

# Plot histogram
plt.hist(data, bins=15, color="skyblue", edgecolor="black", alpha=0.7)
plt.title(f"Negative Hypergeometric Distribution (Skewness: {data_skewness:.2f})")
plt.xlabel("Number of Failures")
plt.ylabel("Frequency")
plt.show()

# Print skewness
print(f"Skewness of the data: {data_skewness:.2f}")
```

## Various Continous distributions in Statistics that are useful in Data Analysis and Data Science

Continous Uniform Distribution:The Continuous Uniform Distribution is a probability distribution where all outcomes within a certain interval are equally likely to occur.

Ex: Used in computer simulations and games to generate random numbers within a specified range

```python
#Generating data from Continous uniform distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Parameters for the uniform distribution
a = 0   # Lower bound
b = 10  # Upper bound
size = 10000  # Number of samples

# Generate data
data = np.random.uniform(a, b, size)

# Calculate skewness
data_skewness = skew(data)

# Plot histogram
plt.hist(data, bins=30, color="skyblue", edgecolor="black", alpha=0.7, density=True)
plt.title(f"Continuous Uniform Distribution (Skewness: {data_skewness:.2f})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Print skewness
print(f"Skewness of the data: {data_skewness:.2f}")
```

Exponential Distribution: The Exponential Distribution is a continuous probability distribution that models the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.

Ex: Survival Analysis; It is used in medical research to model the time to an event such as death or recovery.

```python
#Generating data from exponential distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.exponential(scale=2,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Beta Distribution:The Beta Distribution is a continuous probability distribution defined on the interval
[0,1] and is often used to model random variables that represent proportions or probabilities. It is highly flexible and can take on a variety of shapes, depending on its two parameters α and β. The Beta Distribution is widely used in fields such as Bayesian statistics, project management (PERT distribution), and modeling the distribution of proportions or percentages.

```python
##Generating data from beta  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.beta(a=2,b=5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Gamma Distribution: The Gamma Distribution is a continuous probability distribution that is commonly used to model waiting times or the time until a specific number of events occur in a Poisson process. It generalizes the Exponential Distribution and is often used in fields such as reliability analysis, queuing theory, and Bayesian statistics. The Gamma Distribution has two parameters: a shape parameter (α) and a rate parameter (λ).



```python
##Generating data from gamma distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.gamma(shape=2,scale=5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```

Normal Distribution: The Normal Distribution (also known as the Gaussian Distribution) is one of the most important and widely used probability distributions in statistics. It is a continuous probability distribution that is symmetric around its mean, with its shape resembling a bell curve. Many natural phenomena, measurement errors, and various statistical methods rely on the properties of the normal distribution.

```python
####Generating data from Normal distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

# Generate data for normal distribution
mu, sigma = 0, 0.1  # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Create the histogram
count, bins, ignored = plt.hist(s, 30, density=True)

# Plot the PDF
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Display the plot
plt.show()
```

Chi-Square Distribution: Chi-Square Distribution is a special case of the Gamma Distribution and is widely used in hypothesis testing, particularly in tests of independence, goodness-of-fit tests, and confidence interval estimation for variances. It is a continuous probability distribution with a single parameter: the degrees of freedom (k).

```python
##Generating data from chi-square  distribution and visually representing its skewnwss
import numpy as np
import matplotlib.pyplot as plt

data=np.random.chisquare(df=5,size=1000)
plt.data=sns.histplot(data,kde=True)
plt.show()
```
